## Pré-Experimento
<a id="id1"></a> <br> 
### Implementação de funções auxiliares

```python
# Retorna a média dos valores contindos em `array`
def mean(array):
    return sum(array)/len(array)


# Retorna o desvio padrão dos valores contindos em `array`
def stddev(array):
    _mean = mean(array)
    summation = 0
    
    for num in array:
        summation += (num - _mean)**2
        
    return (summation/len(array))**0.5
```

```python
# Desenha o gráfico para cada solução encontrada
def plot_solution(X, Y, weights, l, s):
    X1 = X[:,0]
    X2 = X[:,1]
    
    get_line_y = lambda x, c0, c1: c0 + c1 * x
    fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(17,20))
    weights_list = random.choices(weights, k=6)
    for index, weights in enumerate(weights_list):
        c0 = weights[0]/weights[2]
        c1 = -weights[1]/weights[2]

        lineX = np.array([X1.min(), X1.max()])
        lineY = np.array([get_line_y(X1.min(), c0, c1), get_line_y(X1.max(), c0, c1)])

        binary_cmap = ListedColormap(['red', 'blue'])
        
        
        ax[int(np.floor(index/2))][index%2].set(xlabel='Característica 1 (x1)', ylabel='Característica 2 (x2)',
               title='Dados de entrada e solução para η = %s e I = (-%s, +%s)' %(str(l), str(s), str(s)))

        scatter = ax[int(np.floor(index/2))][index%2].scatter(X1, X2, c=Y, marker=".", cmap=binary_cmap, linewidth=0)
        ax[int(np.floor(index/2))][index%2].plot(lineX, lineY, color="black", alpha= 0.5)

        legend1 = ax[int(np.floor(index/2))][index%2].legend(*scatter.legend_elements(),
                            loc="upper left", title="Saída", bbox_to_anchor=(1.0, 0.5, 0.5, 0.5))
```

```python
# Realiza a tarefa de experimento
# X: conjunto de treinamento
# Y: repostas supervisonadas
# fit_info: armazena as informações de cara experimento
# learn: taxa de aprendizado para o experimento
# samp: amostra para gerar distribuição uniforme
def fit_and_get_data(X, Y, fit_info, learn, samp):
    all_weights = np.array([0, 0, 0])
    for i in range(100):
        p = Perceptron()
        p.fit(X, Y, learning_rate = learn, sampling_range = samp, verbose = False)

        fit_info['η'].append(learn)
        fit_info['I'].append(samp)
        fit_info['Total Ajustes'].append(np.sum(p.epoch_changes))
        fit_info['Total Épocas'].append(len(p.epoch_changes))
        all_weights = np.vstack([all_weights, p.weights])
        
    return fit_info, all_weights[1:]
    
```

### Discusão dos resultados
A partir da tabela, gráficos e demais resultados, podemos concluir as seguintes definições:

###### Definições para η
1. η alto (0.40): converge rapidamente, porém é instável, ou seja, pode encontrar um mínimo rapidamente ou nunca encontrar;
2. η mediano (0.10): características medianas entre alto e baixo;
3. η baixo (0.01): é mais confiável porém converge lentamente, apesar disso não é garantido encontrar um mínimo global sempre.

###### Definições para I
1. I extremamente largo (100.0): tem-se muitos valores distantes de 1.0, fato que o torna um potencial fator de instabilidade para uma entrada qualquer;
2. I mediano (1.0): tem-se valores mais próximos de 1.0, fato que o torna um ótimo sistema de pesos;
3. I estreito (0.5): tem-se valores razoavelmente próximos de 1.0, fato que o torna um bom sistema de pesos;

*As conclusões abaixo são feitas com base nas definições e nos resultados encontrados*

#### `Exp n. 0: `
Em média, foram necessárias apenas **3 épocas**, com **32 ajustes** para encontrar retas de separação **apertadas**. Para esse conjunto de dados pode ser considerado um **resultado razoável**.

---

#### `Exp n. 1:`
Em apenas **4 épocas em média** precisou realizar **64 ajustes em média** para encontrar retas de separação **frouxas**. Há um pouco de lentidão em relação aos demais em troca de confiança. Para esse conjunto de dados pode ser considerado um **ótimo resultado** e essa configuração **tem potencial para garantir o mesmo desempenho** em um outro conjunto com dados.

---

#### `Exp n. 2:` 
Essa configuração precisou realizar **mais de 5000 ajustes em média** em **mais de 170 épocas rodadas em média** para encontrar retas de separação **apertadas**. É a prova de que parâmetros mal combinados podem atrapalhar o treinamento. Para esse conjunto de dados, a configuração é considerada **péssima**.

---

#### `Exp n. 3:` 
Essa confinguração possui um **ótimo desempenho em termos de quantidade de épocas e ajustes** até a convergência, porém as retas de separação encontradas são **apertadas**. Sendo assim, a configuração para esse conjunto de dados é **considerada mediana**, já que há outras configurações que conseguem retas melhores, com um impacto mínimo no número de épocas e ajustes.

---

#### `Exp n. 4: `
Essa confinguração possui um **ótimo desempenho em termos de quantidade de épocas e ajustes** até a convergência e gerou retas de separação **levemente frouxas**. A configuração para esse conjunto de dados é **considera boa**.

---

#### `Exp n. 5:` 
A configuração gerou **ótimas retas de saparação** em poucos ajudes e épocas, pois, pelo desviu padrão de ambos, percebemos que 
alguns pesos levam a consumir mais tempo enquanto outros levam à rápida convergência. Para esse conjunto de dados pode ser considerado um **ótimo resultado**.

---

#### `Exp n. 6:`
Esse experimento carrega uma boa combinação de parâmetros e seus **números até a convergência são baixos**, porém as retas geradas são **apertadas**. Para esse conjunto de dados pode ser considerado um **resultado razoável para ruim**.

---

#### `Exp n. 7:`
Esse experimento carrega uma boa combinação de parâmetros e seus **números até a convergência são baixos**, porém as retas geradas são **apertadas**. Para esse conjunto de dados pode ser considerado um **resultado razoável para ruim**.

---

#### `Exp n. 8: `
Em **7 épocas em média** precisou realizar **147 ajustes em média** para encontrar retas de separação **apertadas**. Para esse conjunto de dados pode ser considerado um **resultado ruim**.
